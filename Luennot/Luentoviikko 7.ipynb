### Viikon keskeiset asiat

Visuaalinen analytiikka ei ole joukko tekniikoita tai työkaluja vaan pikemminkin analyyttistä päättelyä, jossa visuaaliset menetelmät ovat keskiössä. Sen avulla voidaan hallita ja ykisnkertaistaa suurta määrää informaatiota. Visualisoitua dataa voidaan käyttää päätöksenteon tukena ja datasta esitettävien kysmysten tarkentamiseksi.

Visuaalisen analytiikan prosessi on hyvin samanlainen verrattuna vastaaviin datantieteen iteratiivisiin malleihin. Painopisteenä on kuitenkin visuaalinen puoli. Prosessin on oltava riittävän dynaaminen, sillä jokainen projekti on omanlainen. Visualisoinnin ratkaisut on siis muotoiltava vastaamaan kunkin projektin ainestoa ja keskeisiä ongelmia. Visuaalisen kommunikoinnin tavoitteena on lopulta tuottaa tietämystä esimerkiksi tietyn organisaation toiminnasta, joten visualisoinnissa tulee myös huomioida kohdeyleisö.

Huonoilla visuaalisilla ratkaisuilla ei saada esitettyä haluttua viestiä tarpeeksi selkeästi. Kaikki kaaviot eivät vältämättä sovellu hyvin tietyn datan esittämiseen. Valitun kaavion toimivuus tulee myös varmistaa. Esimerkiksi huonosti skaalautuvat x- tai y-akselit voivat tehdä viestin ymmärtämisestä vähintäänkin haastavaa tai jopa harhaanjohtaa yleisöä.

Yksinkertaiset visuaaliset ratkaisut ovat useimmiten kaikista toimivimpia, jotta olennaisin esitettävä asia saadaan tuotua esille. Niistä yleisö saa viestin nopealla silmäilyllä. Monimutkaisemmilla visualisoinneilla voidaan tehostaa kommunikointia, mutta se usein vaatii yleisöltä pidempää tarkkailua tai tottumusta kyseiseen esitystyyppiin.

### Viisi oivallusta

- Visualisointi kohdistettava oikeaan kohderyhmään

- Yksinkertaiset visualisoinnit toimivia

### Koodidemo

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Haetaan data ja katsotaan mitä se sisältää
url = 'https://raw.githubusercontent.com/pyviz/holoviews/master/examples/assets/station_info.csv'
data = pd.read_csv(url, index_col=0, parse_dates=[0])
data.head()
```

```python
# Mallinnetaan sarakkeen services sisältö seaborn kirjaston avulla pylväsdiagrammiksi
sns.displot(data['services'])
```

```python
# Käytetään seaborn kirjastoa ja mallinnetaan services ja ridership muuttujien välistä
# käyttäytymistä pistekuvion avulla
fig = plt.subplots(figsize=(12,12))
sns.regplot(x='services',y='ridership',data=data)
```
